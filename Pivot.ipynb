# Pivot & Pivot_Table

```python
import pandas as pd
```

```python
df1 = pd.read_csv("weather.csv")
```

```python
df1
```

```python
df1.pivot(index="date",columns="city")
```

```python
df1.pivot(index="date",columns="city",values="humidity")
```

```python
df1.pivot(index="date",columns="city",values="temperature")
```

```python
df1.pivot(index="humidity",columns="city")
```

```python
df2 = pd.read_csv("weather2.csv")
```

```python
df2
```

```python
df2.pivot_table(index="date",columns="city")
```

```python
df2.pivot_table(index="date",columns="city",aggfunc="sum")
```

```python
df2.pivot_table(index="date",columns="city",aggfunc="diff")
```

```python
df2.pivot_table(index="date",columns="city",aggfunc="count")
```

```python
df2.pivot_table(index="date",columns="city",aggfunc="mean")
```

```python
df2.pivot_table(index="date",columns="city",margins=True)
```

```python
df = pd.DataFrame({'foo': ['one', 'one', 'one', 'two', 'two',
                           'two'],
                   'bar': ['A', 'B', 'C', 'A', 'B', 'C'],
                   'baz': [1, 2, 3, 4, 5, 6],
                   'zoo': ['x', 'y', 'z', 'q', 'w', 't']})
```

```python
df
```

```python
df.pivot(index='foo', columns='bar', values='baz')
```

```python
df.pivot(index='foo', columns='bar')
```

```python
df.pivot(index='foo', columns='bar')['baz']
```

```python
df.pivot(index='foo', columns='bar', values=['baz', 'zoo'])
```

```python
df = pd.DataFrame({
       "lev1": [1, 1, 1, 2, 2, 2],
       "lev2": [1, 1, 2, 1, 1, 2],
       "lev3": [1, 2, 1, 2, 1, 2],
       "lev4": [1, 2, 3, 4, 5, 6],
       "values": [0, 1, 2, 3, 4, 5]})
df
```

```python
df.pivot(index="lev1", columns=["lev2", "lev3"],values="values")
```

```python
df.pivot(index=["lev1", "lev2"], columns=["lev3"],values="values")
```
